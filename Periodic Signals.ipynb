```python
import numpy as np 
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 20, 100)
sin_signal = np.sin(x)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,sin_signal, label="Sin(x)")
#plt.plot(x, Cos, label="Cos(x)")
plt.xlim(0, max(x))
plt.title("Periodic Signals Imputation", fontsize= 20)
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize = 15)
plt.grid()
plt.legend()
plt.show()
```

```python


def function(signal, t=0.05):
    index = t * len(signal)
    print(index)
    Range = list(range(len(signal)))
    new_index = random.sample(Range, int(index))
    print(new_index)
    for i in new_index:
        print(i)
        
        signal[i]=np.nan
    return signal
```

```python
new_signal= function(signal=sin_signal, t=0.01)
print(len(new_signal))


plt.figure(num=1, figsize=[6,5], dpi=80, clear=True)
plt.plot(new_signal)
# plt.close()
```

```python
new_signal= function(signal=sin_signal, t=0.02)
print(len(new_signal))


plt.figure(num=1, figsize=[6,5], dpi=80, clear=True)
plt.plot(new_signal)
# plt.close()
```
